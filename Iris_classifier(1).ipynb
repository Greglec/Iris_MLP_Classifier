```python
import sklearn.neural_network as nn
import sklearn.datasets as data
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#upload Iris data
dataset = data.load_iris()
```

```python
#Set input/output
X = dataset.data[:,2:]
Y = dataset.target
Labels = dataset.values()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)
```

```python
#list of neural network parameters to be tested with confusion matrix
param = [{'name':'sgd','param':(5,2), 'a':0.0001},
           {'name':'sgd','param':(5,2), 'a':0.001},
           {'name':'sgd','param':(5,2), 'a':0.01},
           {'name':'sgd','param':(5,2), 'a':0.1},
           {'name':'sgd','param':(15,), 'a':0.0001},
           {'name':'exp22','param':(15,), 'a':0.001},
           {'name':'exp23','param':(15,), 'a':0.01},
           {'name':'exp24','param':(15,), 'a':0.1},
           {'name':'sgd','param':(10,), 'a':0.0001},
           {'name':'exp22','param':(10,), 'a':0.001},
           {'name':'exp23','param':(10,), 'a':0.01},
           {'name':'exp24','param':(10,), 'a':0.1},
           {'name':'exp31','param':(1,1), 'a':0.0001},
           {'name':'exp32','param':(1,1), 'a':0.001},
           {'name':'exp33','param':(1,1), 'a':0.01},
           {'name':'exp34','param':(1,1), 'a':0.1},
           {'name':'exp311','param':(2,2), 'a':0.0001},
           {'name':'exp321','param':(2,2), 'a':0.001},
           {'name':'exp331','param':(2,2), 'a':0.01},
           {'name':'exp341','param':(2,2), 'a':0.1},
           {'name':'exp31','param':(10,10), 'a':0.0001},
           {'name':'exp32','param':(10,10), 'a':0.001},
           {'name':'exp33','param':(10,10), 'a':0.01},
           {'name':'exp34','param':(10,10), 'a':0.1},
           {'name':'exp31','param':(10,2), 'a':0.0001},
           {'name':'exp32','param':(10,2), 'a':0.001},
           {'name':'exp33','param':(10,2), 'a':0.01},
           {'name':'exp34','param':(10,2), 'a':0.1},
    ]

from sklearn.metrics import classification_report,confusion_matrix
plot_list_test = []
for p in param:
    
    clf = nn.MLPClassifier(solver='lbfgs', hidden_layer_sizes=p['param'], activation = 'logistic', alpha=p['a'])
    
    clf.fit(X_train, Y_train)
    
    predict = clf.predict(X_test)
    plot_list_test.append(clf.score(X_train,Y_train))
    print(p['param'], p['a'], '***',clf.score(X_train,Y_train),'-----', clf.score(X_test,Y_test))
    print(confusion_matrix(Y_test,predict))
    print(classification_report(Y_test,predict))

plt.plot(plot_list_test)
plt.show()
```

```python
for p in param:
    
    clf = nn.MLPClassifier(solver='lbfgs', hidden_layer_sizes=p['param'], activation = 'relu', alpha=p['a'])
    
    clf.fit(X_train, Y_train)
    
    predict = clf.predict(X_test)
    plot_list_test.append(clf.score(X_train,Y_train))
    print(p['param'], p['a'], '***',clf.score(X_train,Y_train),'-----', clf.score(X_test,Y_test))
    print(confusion_matrix(Y_test,predict))
    print(classification_report(Y_test,predict))

plt.plot(plot_list_test)
plt.show()
```

```python
#let's plot the decisions borders
clf = nn.MLPClassifier(solver='lbfgs', hidden_layer_sizes=(5,2), activation = 'logistic', alpha=0.0001)
clf.fit(X_train, Y_train)
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

f, axarr = plt.subplots(1, 1, sharey='row', figsize=(10, 7))

axarr.contourf(xx, yy, Z, alpha=0.7)
axarr.scatter(X[:, 0], X[:, 1], c=Y, s=20, edgecolor='black')
```

```python
#prediction error on test data
sum(clf.predict(X_test) == Y_test)/Y_test.shape[0]
```
